```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.__version__
%matplotlib inline


x = np.arange(100)

y1 = np.linspace(0, 1, len(x1)).cumsum()
y2 = np.linspace(0, 1, len(x1)).cumsum()
plt.plot(x1, y1, "b-", x1, y2, "k--")
# plt.plot(x1, np.c_[y1, y2]) second way

plt.title("just a frame to copy")
plt.xlabel("y axis")
plt.ylabel("x axis")
plt.text(20, 40, "point is here") # x, y 
plt.axis([x[0], x[-1], y1[0], y1[-1]]) # xmin, xmax, ymin, ymax !
#plt.legend() # no handle for it, but works
plt.show()
```
